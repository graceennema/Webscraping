# NRF 2025
## Retails Big Show
### January 12-14, 2025
Jacob K. Javits Convention Center

Contains: Company, booth number, website, city, state, country

```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
import re
```

```python
r = requests.get("https://events.nrf.com/annual2025/Public/EventMap.aspx?shMode=E&ID=28009")
```

```python
soup = BeautifulSoup(r.text, "html.parser")
```

```python
companies = soup.findAll("a", {"class" : "exhibitorName"})
```

```python
data = []
for company in companies:

    url = "https://events.nrf.com/annual2025/Public/" + company['href']
    r2 = requests.get(url)
    soup2 = BeautifulSoup(r2.text, "html.parser")
    
    try:
        city = soup2.find("span", {"class" : "BoothContactCity"}).text.strip().rstrip(',')
    except AttributeError:
        city = "N/A"
    
    try:
        state = soup2.find("span", {"class" : "BoothContactState"}).text.strip()
    except AttributeError:
        state = "N/A"
    
    try:
        country = soup2.find("span", {"class" : "BoothContactCountry"}).text.strip()
    except AttributeError:
        country = "N/A"
    
    try:
        site = soup2.find("span", {"class" : "BoothContactUrl"}).find("a").text.strip()
    except AttributeError:
        site = "N/A"
    
    
    a_tag = soup2.find("a", {"class": "aa-mapIt"})
    on_click_attr = a_tag["onclick"]
    booth_numbers = re.findall(r"'(.*?)'", on_click_attr)[-1]
    
    data.append({
        'Company Name' : company.text,
        'Booth' : booth_numbers,
        'Website' : site,
        'City' : city,
        'State' : state,
        'Country' : country
    })
```

```python
df = pd.DataFrame(data)
df
```

```python
df.to_excel('NRF_2025.xlsx', index=False)
```
